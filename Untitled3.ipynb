```python
import pandas as pd
import sqlite3

# 1. ƒê·ªåC FILE V√Ä X·ª¨ L√ù T√äN C·ªòT (QUAN TR·ªåNG)
try:
    df = pd.read_csv('train.csv')
    print("--- ƒê·ªåC FILE TH√ÄNH C√îNG ---")

    # B∆∞·ªõc s·ª≠a l·ªói quan tr·ªçng: X√≥a kho·∫£ng tr·∫Øng th·ª´a ·ªü t√™n c·ªôt
    df.columns = df.columns.str.strip()

    # In ra danh s√°ch c·ªôt ƒë·ªÉ ki·ªÉm tra
    print("C√°c c·ªôt c√≥ trong file:", list(df.columns))

    # Ki·ªÉm tra xem c√≥ c·ªôt Order Date kh√¥ng, n·∫øu kh√¥ng th√¨ th·ª≠ t√¨m c·ªôt t∆∞∆°ng t·ª±
    if 'Order Date' not in df.columns:
        print("‚ö† C·∫¢NH B√ÅO: Kh√¥ng th·∫•y c·ªôt 'Order Date'. ƒêang t√¨m c·ªôt ng√†y th√°ng kh√°c...")
        # Th·ª≠ t√¨m c·ªôt c√≥ ch·ªØ 'Date'
        possible_date_cols = [c for c in df.columns if 'Date' in c]
        print(f"G·ª£i √Ω: C√≥ th·ªÉ c·ªôt ng√†y l√†: {possible_date_cols}")
        # N·∫øu t√¨m th·∫•y, t·ª± ƒë·ªông ƒë·ªïi t√™n c·ªôt ƒë·∫ßu ti√™n t√¨m th·∫•y th√†nh 'Order Date'
        if possible_date_cols:
            old_col = possible_date_cols[0]
            df.rename(columns={old_col: 'Order Date'}, inplace=True)
            print(f"--> ƒê√£ t·ª± ƒë·ªông ƒë·ªïi t√™n c·ªôt '{old_col}' th√†nh 'Order Date' ƒë·ªÉ ch·∫°y ti·∫øp.")

except FileNotFoundError:
    print("‚ùå L·ªñI: Kh√¥ng t√¨m th·∫•y file 'train.csv'. B·∫°n h√£y ki·ªÉm tra l·∫°i t√™n file ƒë√£ upload b√™n c·ªôt tr√°i.")
    df = pd.DataFrame() # T·∫°o b·∫£ng r·ªóng ƒë·ªÉ tr√°nh l·ªói code b√™n d∆∞·ªõi

# 2. N·∫æU ƒê√É C√ì D·ªÆ LI·ªÜU, TI·∫æP T·ª§C X·ª¨ L√ù
if not df.empty and 'Order Date' in df.columns:
    # Chuy·ªÉn ƒë·ªïi ng√†y th√°ng
    df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True, errors='coerce')
    df['Postal Code'] = df['Postal Code'].fillna(0).astype(int)

    # --- T·∫†O DIMENSIONS ---
    # Dim Time
    dim_time = pd.DataFrame()
    dim_time['Order Date'] = df['Order Date'].unique()
    dim_time = dim_time.dropna().sort_values('Order Date')
    dim_time['Date_ID'] = dim_time['Order Date'].dt.strftime('%Y%m%d').astype(int)
    dim_time['Year'] = dim_time['Order Date'].dt.year
    dim_time['Quarter'] = dim_time['Order Date'].dt.quarter
    dim_time['Month'] = dim_time['Order Date'].dt.month

    # Dim Product & Customer & Location
    dim_product = df[['Product ID', 'Category', 'Sub-Category', 'Product Name']].drop_duplicates(subset=['Product ID'])
    dim_customer = df[['Customer ID', 'Customer Name', 'Segment']].drop_duplicates()
    dim_location = df[['Country', 'Region', 'State', 'City', 'Postal Code']].drop_duplicates().reset_index(drop=True)
    dim_location['Location_ID'] = dim_location.index + 1

    # --- T·∫†O FACT TABLE ---
    fact_sales = df.merge(dim_time[['Order Date', 'Date_ID']], on='Order Date', how='left') \
                   .merge(dim_location, on=['Country', 'Region', 'State', 'City', 'Postal Code'], how='left')

    fact_sales = fact_sales[['Row ID', 'Date_ID', 'Customer ID', 'Product ID', 'Location_ID', 'Sales']]

    # --- L∆ØU DATABASE ---
    conn = sqlite3.connect('Train_DW.db')
    fact_sales.to_sql('Fact_Sales', conn, if_exists='replace', index=False)
    dim_time.to_sql('Dim_Time', conn, if_exists='replace', index=False)
    dim_product.to_sql('Dim_Product', conn, if_exists='replace', index=False)
    dim_location.to_sql('Dim_Location', conn, if_exists='replace', index=False)

    print("\n‚úÖ CH√öC M·ª™NG! ƒê√É S·ª¨A L·ªñI V√Ä CH·∫†Y TH√ÄNH C√îNG.")
    print(f"D·ªØ li·ªáu Fact Sales: {len(fact_sales)} d√≤ng.")
else:
    print("\n‚ùå CH∆ØA CH·∫†Y ƒê∆Ø·ª¢C: Vui l√≤ng ki·ªÉm tra l·∫°i t√™n c·ªôt trong file Excel c·ªßa b·∫°n.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# H√†m ch·∫°y SQL ti·ªán l·ª£i
def run_query(sql):
    return pd.read_sql(sql, conn)

# ============================================
# B√ÅO C√ÅO 1: PH√ÇN T√çCH DOANH S·ªê THEO NƒÇM (XU H∆Ø·ªöNG)
# ============================================
print("--- üìä B√ÅO C√ÅO 1: DOANH S·ªê THEO C√ÅC NƒÇM ---")
query_year = """
SELECT
    t.Year,
    SUM(f.Sales) as Doanh_Thu
FROM Fact_Sales f
JOIN Dim_Time t ON f.Date_ID = t.Date_ID
GROUP BY t.Year
ORDER BY t.Year
"""
df_year = run_query(query_year)
display(df_year)

# V·∫Ω bi·ªÉu ƒë·ªì c·ªôt cho Doanh s·ªë theo nƒÉm
plt.figure(figsize=(10, 5))
sns.barplot(data=df_year, x='Year', y='Doanh_Thu', palette='viridis')
plt.title('T·ªîNG DOANH THU THEO NƒÇM (Trend Analysis)', fontsize=14)
plt.ylabel('Doanh thu (USD)')
plt.show()

# ============================================
# B√ÅO C√ÅO 2: TOP 5 BANG (STATE) MUA NHI·ªÄU NH·∫§T
# ============================================
print("\n--- üåé B√ÅO C√ÅO 2: TOP 5 BANG C√ì DOANH S·ªê CAO NH·∫§T ---")
query_state = """
SELECT
    l.State,
    SUM(f.Sales) as Doanh_Thu
FROM Fact_Sales f
JOIN Dim_Location l ON f.Location_ID = l.Location_ID
GROUP BY l.State
ORDER BY Doanh_Thu DESC
LIMIT 5
"""
df_state = run_query(query_state)
display(df_state)

# V·∫Ω bi·ªÉu ƒë·ªì tr√≤n (Pie Chart)
plt.figure(figsize=(8, 8))
plt.pie(df_state['Doanh_Thu'], labels=df_state['State'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('TOP 5 TH·ªä TR∆Ø·ªúNG L·ªöN NH·∫§T', fontsize=14)
plt.show()

# ============================================
# B√ÅO C√ÅO 3: PH√ÇN T√çCH CHI TI·∫æT DANH M·ª§C (Drill-Down)
# ============================================
print("\n--- üì¶ B√ÅO C√ÅO 3: DOANH S·ªê THEO DANH M·ª§C S·∫¢N PH·∫®M ---")
query_cat = """
SELECT
    p.Category,
    SUM(f.Sales) as Doanh_Thu
FROM Fact_Sales f
JOIN Dim_Product p ON f."Product ID" = p."Product ID"
GROUP BY p.Category
ORDER BY Doanh_Thu DESC
"""
df_cat = run_query(query_cat)
display(df_cat)
```
